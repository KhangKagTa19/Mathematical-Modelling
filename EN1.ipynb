```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
%matplotlib inline
```

```python
t = np.linspace(0, 4, 100)
#change function
R = ((-2 + 0*np.sqrt(10))/1)*(np.e**((3 + 0*np.sqrt(10))*t/1))*(np.sqrt(2)*np.sin(np.sqrt(2)*t)) + ((-np.sqrt(2) + 0*np.sqrt(10))/1)*(np.e**((3 + 0*np.sqrt(10))*t/1))*(-np.sqrt(2)*np.cos(np.sqrt(2)*t))
J = ((-2 + 0*np.sqrt(10))/1)*(np.e**((3 + 0*np.sqrt(10))*t/1))*(1*np.cos(np.sqrt(2)*t)) + ((-np.sqrt(2) + 0*np.sqrt(10))/1)*(np.e**((3 + 0*np.sqrt(10))*t/1))*(np.sin(np.sqrt(2)*t))

#
plt.figure(figsize = (7, 7))
font = font_manager.FontProperties(family='Arial',
                                   weight='normal',
                                   style='normal', size=14)
plt.plot(t, R)
plt.plot(t, J)

#change title
plt.title("LOVE BETWEEN A EAGER BEAVER\nAND NARCISSISTIC NERD")
#
plt.xlabel("Time", {'size':14})
plt.ylabel("Love for the other", {'size':14})
plt.legend(["Romeo's", "Juliet's"], loc = "best", prop=font)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.plot(R, J, color = "red")
plt.xlim([-4.4, 4.4])
plt.ylim([-4.4, 4.4])

R = np.linspace(-4.4, 4.4, 9)
#change nullcline
plt.plot(R, R*-3/2, linestyle = "--", color = "blue")
plt.plot(R,  R*1/3, linestyle = "--", color = "purple")

R, J = np.meshgrid(np.linspace(-4, 4, 9),np.linspace(-4, 4, 9))
#change a, b, c, d
_R = 3 * R + 2 * J
_J = -1 * R + 3 * J

M = (np.hypot(_R, _J))
M[M == 0] = 1
_R /= M
_J /= M

#change title
plt.title("LOVE BETWEEN A EAGER BEAVER\nAND NARCISSISTIC NERD")
#
plt.xlabel("Romeo's love for Juliet", {'size':14})
plt.ylabel("Juliet's love for Romeo", {'size':14})
plt.quiver(R, J, _R, _J, color = 'g', width = 0.005)
plt.streamplot(R, J, _R, _J, density = 0.2, linewidth = 1, color = "gray", broken_streamlines = False)
plt.legend(["Trajectory", "Nullcline 1", "Nullcline 2", "Vector field", "Trajectory"], loc = "upper left")
plt.show()
```
