```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.lines import Line2D
%matplotlib inline
```

```python
t = np.linspace(0, 10, 1000000)
#change function
R = np.power(np.e, -t) * ((-2/3) * (0 * np.cos(np.sqrt(3) * t) - np.sqrt(3) * np.sin(np.sqrt(3) * t)) +(1/np.sqrt(3)) * (np.sqrt(3) * np.cos(np.sqrt(3) * t) + 0 * np.sin(np.sqrt(3) * t)))
J = np.power(np.e, -t) * ((-2/3) * (3 * np.cos(np.sqrt(3) * t) - 0 * np.sin(np.sqrt(3) * t)) +(1/np.sqrt(3)) * (0 * np.cos(np.sqrt(3) * t) + 3 * np.sin(np.sqrt(3) * t)))
#
plt.figure(figsize = (7, 7))
font = font_manager.FontProperties(family='Arial',
                                   weight='normal',
                                   style='normal', size=14)
plt.plot(t, R)
plt.plot(t, J)
#change title
plt.title("LOVE BETWEEN A HERMIT\nAND A CAUTIOUS LOVER")
#
plt.xlabel("Time", {'size':14})
plt.ylabel("Love for the other", {'size':14})
plt.legend(["Romeo's", "Juliet's"], loc = "upper right", prop=font)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.plot(R, J, color = "red")
plt.xlim([-4.4, 4.4])
plt.ylim([-4.4, 4.4])

R = np.linspace(-4.4, 4.4, 9)
plt.plot(R, -R, linestyle = "--", color = "black")
plt.plot(R, 3 * R, linestyle = "--", color = "black")
R, J = np.meshgrid(np.linspace(-4, 4, 9),np.linspace(-4, 4, 9))
#change a, b, c, d
_R = -1 * R + -1 * J
_J = 3 * R + -1 * J

M = (np.hypot(_R, _J))
M[M == 0] = 1
_R /= M
_J /= M

#change title
plt.title("LOVE BETWEEN A HERMIT\nAND A CAUTIOUS LOVER")
#
plt.xlabel("Romeo's love for Juliet", {'size':14})
plt.ylabel("Juliet's love for Romeo", {'size':14})
plt.quiver(R, J, _R, _J, color = 'g', width = 0.005)
plt.streamplot(R, J, _R, _J, density = 0.2, linewidth = 1, color = "gray", broken_streamlines = False)
plt.legend(["Trajectory", "Nullcline 1", "Nullcline 2", "Vector field", "Trajectory"], loc = "upper left")
plt.show()
```
