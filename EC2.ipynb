```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.lines import Line2D
%matplotlib inline
```

```python
t = np.linspace(0, 10, 1000000)
#change function
R = ((4 + np.sqrt(6)) * np.power(np.e, ((-1 + np.sqrt(6)) * t)) + (-4 + np.sqrt(6)) * np.power(np.e, ((-1 - np.sqrt(6)) * t))) / (2 * np.sqrt(6))
J = ((-2 + 2 * np.sqrt(6)) * np.power(np.e, ((-1 + np.sqrt(6)) * t)) + (2 + 2 * np.sqrt(6)) * np.power(np.e, ((-1 - np.sqrt(6)) * t)))/ (2 * np.sqrt(6))
#
plt.figure(figsize = (7, 7))
font = font_manager.FontProperties(family='Arial',
                                   weight='normal',
                                   style='normal', size=14)
plt.plot(t, R)
plt.plot(t, J)
#change title
plt.title("LOVE BETWEEN AN EAGER BEAVER\nAND A CAUTIOUS LOVER")
#
plt.xlabel("Time", {'size':14})
plt.ylabel("Love for the other", {'size':14})
plt.legend(["Romeo's", "Juliet's"], loc = "upper right", prop=font)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.plot(R, J, color = "red")
plt.xlim([-12.3, 12.3])
plt.ylim([-12.3, 12.3])

R = np.linspace(-12.3, 12.3, 9)
plt.plot(R,  -1 * R, linestyle = "--", color = "black")
plt.plot(R, (2/3) *R, linestyle = "--", color = "black")
R, J = np.meshgrid(np.linspace(-12, 12, 9),np.linspace(-12, 12, 9))
#change a, b, c, d
_R = 1 * R + 1 * J
_J = 2 * R + -3 * J

M = (np.hypot(_R, _J))
M[M == 0] = 1
_R /= M
_J /= M

#change title
plt.title("LOVE BETWEEN AN EAGER BEAVER\nAND A CAUTIOUS LOVER")
#
plt.xlabel("Romeo's love for Juliet", {'size':14})
plt.ylabel("Juliet's love for Romeo", {'size':14})
plt.quiver(R, J, _R, _J, color = 'g', width = 0.005)
plt.streamplot(R, J, _R, _J, density = 0.2, linewidth = 1, color = "gray", broken_streamlines = False)
plt.legend(["Trajectory", "Nullcline 1", "Nullcline 2", "Vector field", "Trajectory"], loc = "upper left")
plt.show()
```
